## 54525 

```python
pulse = 54525
PQ1, tQ1 = get_sig(pulse, signals['IC_P_Q1'])
PQ4, tQ4 = get_sig(pulse, signals['IC_P_Q4'])
output = imas_west.get(pulse, 'bolometer')

fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True)
ax1.plot(output.time - 32, output.power_radiated_total/1e3, lw=2, label='Total Rad.Power')
ax2.plot(tQ1, PQ1, lw=2, label='Q1', color='C1')
ax2.plot(tQ4, PQ4, lw=2, label='Q4', color='C2')
ax1.set_xlim(0, 10)
ax2.legend()
[a.set_ylabel('P [kW]', fontsize=14) for a in (ax1, ax2)]
[a.grid(True, alpha=.4) for a in (ax1, ax2)]
ax2.set_xlabel('t [s]', fontsize=14)
fig.subplots_adjust(hspace=.1)
ax1.set_title(f'WEST #{pulse}')
fig.savefig(f'WEST_{pulse}_Prad_vs_IC.png', dpi=150)
```
